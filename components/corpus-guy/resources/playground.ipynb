```python
import subprocess

def run_command(command):
    """
    Run a command and return the output.
    """
    result = subprocess.run(command, shell=True, capture_output=True, text=True)
    if result.returncode != 0:
        raise Exception(f"Command failed: {command}\n{result.stderr}")
    return result.stdout

# copy all crashing seeds to /tmp/corpus-kickstart-crashes/id:<id with 6 digits>_<sha256sum of content>
# run_command("rm -rf /tmp/corpus-kickstart-crashes && mkdir -p /tmp/corpus-kickstart-crashes && i=1 && for file in $(cat ./known-oss-fuzz-crashes); do sha256sum=$(sha256sum $file | cut -d ' ' -f 1); cp $file /tmp/corpus-kickstart-crashes/id:$(printf '%06d' $i)_$sha256sum; i=$((i+1)); done")

num_crashing_seeds = len(run_command("ls /tmp/corpus-kickstart-crashes").split("\n")) - 1
print(f"Number of crashing seeds: {num_crashing_seeds}")

# hash and copy all files from /tmp/AIxCC/ (recursively) in the same way
run_command(f"i={num_crashing_seeds + 1} && for file in $(find /tmp/AIxCC/ -type f); do sha256sum=$(sha256sum $file | cut -d ' ' -f 1); cp $file /tmp/corpus-kickstart-crashes/id:$(printf '%06d' $i)_$sha256sum; i=$((i+1)); done")
num_crashing_seeds = len(run_command("ls /tmp/corpus-kickstart-crashes").split("\n")) - 1
print(f"Number of crashing seeds: {num_crashing_seeds}")
```

```python
# package grammars

from morpheus.magic import ALIAS_TO_NAME
all_file_formats = set(ALIAS_TO_NAME.values())

import glob
import os
import shutil

os.makedirs("/tmp/corpus-grammars", exist_ok=True)
for format in sorted(all_file_formats):
    if any (glob.glob(f"/home/ruaronicola/specler-data/BEST_GRAMMARS_*/{format}.py")):
        print(f"Found {format} in BEST_GRAMMARS")
        grammar_path = glob.glob(f"/home/ruaronicola/specler-data/BEST_GRAMMARS_*/{format}.py")[0]
        shutil.copy(grammar_path, f"/tmp/corpus-grammars/{format}.py")
    elif any (glob.glob(f"/home/ruaronicola/artiphishell/libs/nautilus/grammars/reference/{format}.py")):
        print(f"Found {format} in nautilus")
        grammar_path = glob.glob(f"/home/ruaronicola/artiphishell/libs/nautilus/grammars/reference/{format}.py")[0]
        shutil.copy(grammar_path, f"/tmp/corpus-grammars/{format}.py")
    elif any (glob.glob(f"/home/ruaronicola/artiphishell/components/grammar-composer/playground/grammar_drafts/{format}.py")):
        print(f"Found {format} in grammar-composer")
        grammar_path = glob.glob(f"/home/ruaronicola/artiphishell/components/grammar-composer/playground/grammar_drafts/{format}.py")[0]
        shutil.copy(grammar_path, f"/tmp/corpus-grammars/{format}.py")

    # look for a corpus-filtered dir
    if os.path.isdir(f"/tmp/corpus-filtered/{format.lower()}"):
        print(f"Found {format} in corpus-filtered")
        grammar_str = ""
        for seed in glob.glob(f"/tmp/corpus-filtered/{format.lower()}/*"):
            # encode seed as bytes \x encoded
            with open(seed, "rb") as f:
                content = f.read()
            content_str = "".join([f"\\x{byte:02x}" for byte in content])
            grammar_str += f'ctx.literal("START", b"{content_str}")\n'
        # write grammar str to /tmp/corpus-grammars/{format}@CORPUS.py
        with open(f"/tmp/corpus-grammars/{format}@CORPUS.py", "w") as f:
            f.write(grammar_str)

# rm -rf /tmp/corpus-grammars.tar && tar -C /tmp/corpus-grammars -cf /tmp/corpus-grammars.tar .
```
